```python
# Importing the tools we
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

# Show your data in tabular or other visualization ways
# Data Preprocessing
# Data Splitting
# Write your code here

# Reading and printing the data head
data = pd.read_csv("abalone.data", header=None)
column_names = ['Sex', 'Length', 'Diameter', 'Height', 'Whole_weight', 'Shucked_weight', 'Viscera_weight', 'Shell_weight', 'Rings']
df = pd.DataFrame(data)
df.columns = column_names
print(df.head())

# Making a histogram in the target variable which is the "Rings" variable
df["Rings"].hist()
```

```python
# Making the data correlation
numeric_columns = df.select_dtypes(include='number')
corr = numeric_columns.corr(method="pearson")
cmap = sns.diverging_palette(250, 354, 80, 60, center="dark", as_cmap=True)
sns.heatmap(corr, vmax=1, vmin=-0.5, cmap=cmap, square=True, linewidths=0.2)
plt.show()
```
